```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# initial parameters
initial_price = 100
drift = 0.08
volatility = 0.2
time_period = 1.0
steps = 252
dt = time_period / steps
paths_count = 10
```

```python
# fixing random values
np.random.seed(42)

# creating array to store paths
price_paths = np.zeros((steps + 1, paths_count))
price_paths[0] = initial_price

# generating GBM paths
for day in range(1, steps + 1):
    random_values = np.random.normal(0, 1, paths_count)
    price_paths[day] = price_paths[day - 1] * np.exp(
        (drift - 0.5 * volatility**2) * dt
        + volatility * np.sqrt(dt) * random_values
    )
```

```python
# plotting the simulated paths
plt.figure(figsize=(10, 5))

for j in range(paths_count):
    plt.plot(price_paths[:, j], linewidth=1)

plt.title("Geometric Brownian Motion: Simulated Price Paths")
plt.xlabel("Time Steps (Days)")
plt.ylabel("Price")
plt.show()
```

```python
# extracting final prices from all paths
last_prices = price_paths[-1]

plt.figure(figsize=(7, 5))
plt.hist(last_prices, bins=10, edgecolor='black')

plt.title("Histogram of Final Prices")
plt.xlabel("Final Price")
plt.ylabel("Frequency")

plt.show()
```
